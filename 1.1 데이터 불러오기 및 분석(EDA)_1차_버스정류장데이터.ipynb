# 3.데이터 분석하기
+ KeyPoint : 데이터의 형태를 살펴보고 다양한 분석기법을 통해 모델링에 적합하도록 정제요소를 선별할 수 있다.
  * 데이터들의 패턴 탐색
  * 변수들간의 관계 파악

```python
sns.set(font='AppleGothic', 
        rc={"axes.unicode_minus":False},
        style='darkgrid',
        font_scale = 2)
```

```python
# 구 별로 버스 정류장의 개수

seoul_bus_station_ARS = df.groupby(by=["자치구"], as_index=False)["버스정류장ARS번호"].nunique()
seoul_bus_station_ARS = seoul_bus_station_ARS.sort_values(by = ["버스정류장ARS번호"], ascending = False)
seoul_bus_station_ARS.plot(x= '자치구', kind='bar', figsize=(15,5))
```

#### [실습문제6] 구 별로 버스 노선이 몇 개나 지나가는지 궁금합니다.

* 구별로 버스 노선의 개수를 확인해 주세요

```python
# 지역구별 노선수
seoul_bus_station_line = df.groupby(by=["자치구"], as_index=False)["노선번호"].nunique()
seoul_bus_station_line = seoul_bus_station_line.sort_values(by=["노선번호"],ascending=False)
seoul_bus_station_line.plot(x= '자치구', kind='bar', figsize=(15,5))
```

#### [실습문제7] 각 구별로 승차 총 승객수, 하차 총 승객수를 알아봅시다

* groupby 를 활용해 봅시다.

```python
# 아래에 실습코드를 작성하고 결과를 확인합니다.
seoul_bus_station_sum = df.groupby(by=["자치구"], as_index=False)[["승차총승객수","하차총승객수"]].sum()
seoul_bus_station_sum = seoul_bus_station_sum.set_index('자치구')


# 각 그룹의 외국여자 비율과 외국남자 비율 계산
takeon = seoul_bus_station_sum['승차총승객수']
takeoff = seoul_bus_station_sum['하차총승객수']

# 그래프 그리기
plt.figure(figsize=(15,8))
takeon.plot(kind='bar', color='orange', position=1, width=0.3)
takeoff.plot(kind='bar', color='lightgreen', position=2, width=0.3)
plt.xticks(rotation=45, ha='right')
plt.xlabel('자치구')
plt.ylabel('승객수(명)')
plt.legend(['승차총승객수','하차총승객수'])
plt.show()
```

```python
sns.set(font='AppleGothic', 
        rc={"axes.unicode_minus":False},
        style='darkgrid',
        font_scale = 1)

plt.pie(seoul_bus_station_sum['승차총승객수'], labels=seoul_bus_station_sum.index, autopct='%1.1f%%')
plt.title('각 자치구의 버스 승차총승객수 비율')
plt.show()

# 하차총승객수 파이플롯
plt.pie(seoul_bus_station_sum['하차총승객수'], labels=seoul_bus_station_sum.index, autopct='%1.1f%%')
plt.title('각 자치구의 버스 하차총승객수 비율')
plt.show()
```

#### [실습문제8] 각 구별로 승차 평균 승객수, 하차 평균 승객수를 알아봅시다

```python
# 각 구별로 승차 평균 승객수, 하차 평균 승객수
seoul_bus_station_mean = df.groupby(by=["자치구"], as_index=False)[["승차총승객수","하차총승객수"]].mean()
seoul_bus_station_mean.rename(columns={'승차총승객수':'승차평균승객수','하차총승객수':'하차평균승객수'}, inplace=True)
seoul_bus_station_mean
```

#### [실습문제9] 데이터 프레임을 합쳐보겠습니다. 

```python
# 데이터 프레임 합치기

a = pd.merge(seoul_bus_station_ARS,seoul_bus_station_line,how='inner', on='자치구')
b = pd.merge(a,seoul_bus_station_sum,how='inner', on='자치구')
seoul_bus_station = pd.merge(b,seoul_bus_station_mean,how='inner', on='자치구')
```

```python
# column 이름을 바꿔주겠습니다.
seoul_bus_station.rename(columns={'버스정류장ARS번호':'정류장수','노선번호':'노선수'}, inplace=True)
```
